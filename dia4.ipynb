## Challenge 2 ü§∫ - 20 min

Crear una tarjeta personal. El usuario (mediante `input`) puede elegir:

1. `Crear` datos nuevos
2. `Modificar` datos
3. `Eliminar` datos

```python
# crear una tarjeta
tarjeta = {}

# preguntar 

opcion = input("elija una opcion: ")

if opcion == '1':
    nombre = input("ingrese su nombre: ")
    tarjeta["nombre"] = nombre
    print(tarjeta)

    edad = input("ingrese su edad: ")
    tarjeta["edad"] = edad
    print(tarjeta)

    correo = input("ingrese su correo: ")
    tarjeta["correo"] = correo
    print(tarjeta)

elif opcion == "2":
    clave = input("que dato quiere modificar?: ")
    nuevo_valor = input ("Ingrese el nuevo valor: ")
    tarjeta[clave] = nuevo_valor
    print(tarjeta)

elif opcion == "3":
    clave = input("que dato quiere eliminar?: ")
    tarjeta.pop(clave)
    print(tarjeta)

else :
    print("opcion no valida")



    


```

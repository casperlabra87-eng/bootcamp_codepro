Ejemplo del resultado esperado:

```
['h', 'o', 'l', 'i', 's']

['h', 'a', 'l', 'l', 'o']
```

````python
# crear una funcion

lista = ['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']


def imprimir_lista():
    for i in lista:
        print(i)
        
imprimir_lista()
````
